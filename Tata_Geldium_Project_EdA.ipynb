**Step 5: Data Cleaning**

```python
# prompt: - 'Suggest an imputation strategy for missing Income, Credit_Score, Loan_Balance values based on industry best practices.'

# For numerical columns like Income, Credit_Score, and Loan_Balance,
# median imputation is often a robust strategy as it is less sensitive to outliers
# compared to mean imputation.

# Impute missing 'Income' with the median
median_income = df['Income'].median()
df['Income'].fillna(median_income, inplace=True)

# Impute missing 'Credit_Score' with the median
median_credit_score = df['Credit_Score'].median()
df['Credit_Score'].fillna(median_credit_score, inplace=True)

# Impute missing 'Loan_Balance' with the median
median_loan_balance = df['Loan_Balance'].median()
df['Loan_Balance'].fillna(median_loan_balance, inplace=True)

# Verify that there are no more missing values in these columns
print(df[['Income', 'Credit_Score', 'Loan_Balance']].isnull().sum())
```
